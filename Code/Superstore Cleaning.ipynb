## Gathering Data

Data Overview

1. row_id: Nomor unik untuk setiap baris data (record).
2. order_id: ID unik untuk setiap pesanan (order).
3. order_date: Tanggal saat pesanan dilakukan oleh pelanggan.
4. ship_date: Tanggal pengiriman dilakukan oleh toko.
5. ship_mode: Metode pengiriman yang digunakan.
6. customer_id: ID unik untuk setiap pelanggan.
7. customer_name: Nama pelanggan.
8. segment: Kategori pelanggan berdasarkan jenis pasar.
9. country: Negara tempat pelanggan berada.
10. city: Kota tempat pelanggan berada.
11. state: Negara bagian tempat pelanggan berada.
12. postal_code: Kode pos lokasi pelanggan.
13. region: Wilayah geografis di dalam negara.
14. product_id: ID unik untuk setiap produk.
15. category: Kategori utama produk.
16. subcategory: Subkategori produk.
17. product_name: Nama produk.
18. sales: Total nilai penjualan dari item tersebut.
19. quantity: Jumlah unit produk yang dijual dalam transaksi.
20. discount: ersentase diskon yang diberikan.
21. profit: Keuntungan bersih dari penjualan produk.
